```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Fitness data set new.csv")
df.head()
```

```python
def dynamic_price(row):
    base_price = row["Price (INR)"]
    
    if row["Demand_Ratio"] >= 0.8:
        return base_price * 1.2   # Increase price by 20%
    elif row["Demand_Ratio"] <= 0.4:
        return base_price * 0.85  # Discount
    else:
        return base_price
```

```python
df["Dynamic_Price"] = df.apply(dynamic_price, axis=1)
df[["Price (INR)", "Dynamic_Price", "Demand_Ratio"]].head()
```

```python
df["Original_Revenue"] = df["Price (INR)"] * df["Number Booked"]
df["Dynamic_Revenue"] = df["Dynamic_Price"] * df["Number Booked"]

df[["Original_Revenue", "Dynamic_Revenue"]].sum()
```

#Report
A rule-based dynamic pricing algorithm was developed using demand ratios. 
Prices were increased during high-demand periods and discounted during low-demand periods. Revenue simulations showed potential improvement in 
per-class and overall revenue.
